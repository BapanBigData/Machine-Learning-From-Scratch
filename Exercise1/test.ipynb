```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# tells matplotlib to embed plots within the notebook
%matplotlib inline
```

```python
X = stats.uniform.rvs(20, 100, size=(5, 200))
X.shape
```

```python

sns.set(style="whitegrid")

# Create a histogram of the house sizes
plt.figure(figsize=(10, 6))
sns.histplot(X[:, 0], bins=50, kde=True, color='skyblue')

# Add titles and labels
plt.title('Synthetic House Sizes Following a Log-normal Distribution')
plt.xlabel('House Size (Square Feet) Scaled')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python

def get_scaled_feature(X: np.array):
    m = np.mean(X, axis=0)
    s = np.std(X, axis=0)
    
    return (X - m) / s
```

```python
X_scaled = get_scaled_feature(X)
X_scaled.shape
```

```python

sns.set(style="whitegrid")

# Create a histogram of the house sizes
plt.figure(figsize=(10, 6))
sns.histplot(X_scaled[:, 0], bins=50, kde=True, color='skyblue')

# Add titles and labels
plt.title('Synthetic House Sizes Following a Log-normal Distribution')
plt.xlabel('House Size (Square Feet) Scaled')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
